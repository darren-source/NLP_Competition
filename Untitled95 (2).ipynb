```python
# Create and define a function to clean the text
def clean_data(text):
  text = text.lower()
  text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
  text = re.sub(r'\@\w+|\#', '', text)
  tokens = word_tokenize(text)
  stop_words = set(stopwords.words('english'))
  tokens = [word for word in tokens if word not in stop_words]
  lemmatizer = WordNetLemmatizer()
  tokens = [lemmatizer.lemmatize(word) for word in tokens]
  return ' '.join(tokens)
```

```python
# Apply the cleaning function to the 'text' column of the training data
train['clean_text'] = train['text'].apply(clean_data)
```

**Text Data Cleaning**

Cleaning data is crucial for NLP tasks and creating an accurate model.  Converting all text to lower cases ensure that the algorithms do not treat the same words in different cases as different words. Stop words were removed since they appear frequently in texts and are not useful for text analysis.  Stemming/Lemmatization help to reduce words to their base or root form and thus reducing the number of unique words that model needs to interpret.

Tokenization was used to break down text into smaller units.  These smaller units help to improve model accuracy by breaking text into meaningful units.  Padding is also important to ensure that all text sequences in a dataset are the same length so that they can be processed in batches by a NLP model.

```python
# import tensorflow modules for text processing and neural networkds
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# Initialize a tokenizer to convert text sequences to integer sequences
tokenizer = Tokenizer(num_words=1000)
texts = train['clean_text']
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
```

```python
# Pad the sequence so that all texts have the same length
padded_sequences = pad_sequences(sequences, maxlen=10)
```

```python
# Import models for building a Sequential model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
# Download the GloVe pre-trained work embedings model from the Stanford NLP Group website
!wget http://nlp.stanford.edu/data/glove.6B.zip
```

```python
# Unzip the downloaded GloVe file
!unzip glove.6B.zip
```

```python
# Define a function to load GloVe embeddings from a file into a dictionary
def load_glove_model(file):
    print("Loading GloVe Model")
    glove_model = {}
    with open(file, 'r', encoding='utf8') as f:
        for line in f:
            split_line = line.split()
            word = split_line[0]
            embedding = np.array([float(val) for val in split_line[1:]])
            glove_model[word] = embedding
    print(f"Done. {len(glove_model)} words loaded!")
    return glove_model

# Specify the path to the specific GloVe file
glove_path = 'glove.6B.100d.txt'
glove_vectors = load_glove_model(glove_path)
```

```python
# Determine the vocabulary size
vocab_size = len(tokenizer.word_index) + 1
embedding_dim = 100

# Intialize an embedding matrix to zero
embedding_matrix = np.zeros((vocab_size, embedding_dim))
for word, i in tokenizer.word_index.items():
    if i < vocab_size:  # Adjust based on your vocabulary size
        embedding_vector = glove_vectors.get(word)
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
```

**Model Architecture**

The first part of the model was vectorization.  Vectorization is a fundamental step in preparing text data for NLP machine learning.  GloVe(Global Vectors for Word Representation) was used because it has signifcant advantages, especially in improving the accuracy of models.  GloVe embeddings are trained on massive corpora containing billions of tokens.  This extensive training allows GloVe to capture deep semantic and syntactic meaning of words.

My first model is structured to classify text into binary categories.  The input layer takes sequences of tokens as inputs.  The next layer is an embeding layer.  This layer uses the GloVe pretrained embedding matrix.  The global average pooling 1D layer reduces the dimensionality of the data by averaging the sequence dimension.  The dense layer is a fully connected layer that applies a Relu function activation function to introduce non-linearities into the model, enabling the model to learn complex patterns.  Finally, the output layer is a single neuron suitable for binary classification.  This sigmoid ffunction output a a probability indicating the likelihood of the input being classified as a disaster tweet.

An SGD optimizer was chosen for its ability to handle large datasets and high dimensionality.  A Binary Cross-entropy Loss function was used since the problen was a binary classification problem.

```python
# Define the models input layer
inputs = tf.keras.Input(shape=(None,), dtype='int64')
x = tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim, weights=[embedding_matrix])(inputs)  # Embedding layer
x = tf.keras.layers.GlobalAveragePooling1D()(x)
x = tf.keras.layers.Dense(16, activation='relu')(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)
model = tf.keras.Model(inputs=inputs, outputs=outputs)
```

```python
# Compile the model with SGD optimizer, and binary cross-entropy, and accuracy metrics
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# fit the model on the padded sequence of text
history = model.fit(padded_sequences, train['target'], epochs=5, batch_size=32)
```

```python
# Print a model summary
model.summary()
```

```python
# Plot 2 graphs, training accuracy and training loss
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].plot(history.history['accuracy'])
axs[0].set_title('Updated Model Accuracy')
axs[0].set_ylabel('Accuracy')
axs[0].set_xlabel('Epoch')
axs[0].legend(['Train', 'Validation'], loc='upper left')

axs[1].plot(history.history['loss'])
axs[1].set_title('Model Loss')
axs[1].set_ylabel('Loss')
axs[1].set_xlabel('Epoch')
axs[1].legend(['Train', 'Validation'], loc='upper left')
plt.tight_layout()
plt.show()
```

**Improved Model**

The revised model architecture incorporates a Bideirectional LSTM layer and a drop-out layer along with a different optimizer.  A Bidirectional LSTM layer has the ability to process data in both a forward and reverse direction.  This dual processing can process contextual information from the past and the future, leading to a better understanding of the context of the text.  The dropout layer is important in a complex model as it tends to prevent over fitting.  The Adam optimizeris is an algorithm that can handle sparse gradients on noisy problems, this is common in NLP tasks.

```python
# Define a new model with a Bidirectional LSTM layer
inputs = tf.keras.Input(shape=(None,), dtype='int64')
x = tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim, weights=[embedding_matrix])(inputs)
x = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32))(x)
x = tf.keras.layers.Dropout(0.5)(x)
x = tf.keras.layers.Dense(16, activation='relu')(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)
model2 = tf.keras.Model(inputs=inputs, outputs=outputs)
```

```python
# Compile a second, improved model with an Adam optimizer and binary cross-entropy loss function
model2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model on the padded sequences and target labels
history2 = model2.fit(padded_sequences, train['target'], epochs=10, batch_size=32)
```

```python
# Print a summary
model2.summary()
```

```python
# Create 2 graphs, one graph for plotting accuracy and the other for loss
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].plot(history2.history['accuracy'])
axs[0].set_title('Updated Model Accuracy')
axs[0].set_ylabel('Accuracy')
axs[0].set_xlabel('Epoch')
axs[0].legend(['Train', 'Validation'], loc='upper left')

axs[1].plot(history2.history['loss'])
axs[1].set_title('Model Loss')
axs[1].set_ylabel('Loss')
axs[1].set_xlabel('Epoch')
axs[1].legend(['Train', 'Validation'], loc='upper left')
plt.tight_layout()
plt.show()
```
